## Ejercicio 1 

Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulación para encontrar la probabilidad estimada de que la puntuación sea mayor a 7 o un número par. Una simulación es una repetición del mismo experimento multiples veces para observar su comportamiento:

- Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces, y suma el número de ambos dados).
- Lleva ua cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
- Divide el número del paso 2 entre el número de iteraciones(1000).

```python
# Todo
import numpy as np

# Definimos la función de lanzamiento de dos dados y obtener la suma
def lanzamientos_dados():
    dado1 = np.random.randint(1, 7)
    dado2 = np.random.randint(1, 7)
    return dado1 + dado2

# Contador para los resultados
contador_mayor_7 = 0
contador_par = 0

num_experimentos = 1000

# Simulación
for _ in range(num_experimentos):
    resultado = lanzamientos_dados()
    if resultado > 7:
        contador_mayor_7 += 1
    if resultado % 2 == 0:
        contador_par += 1

# Calculamos la probabilidad
prob_may_7 = contador_mayor_7 / num_experimentos
prob_num_par = contador_par / num_experimentos

# Hacemos el print
print("Probabilidad de obtener un numero mayor a 7:", prob_may_7)
print("Probabilidad de obtener un numero par:", prob_num_par)
```

## Ejercicio 2

Una caja contiene 10 bolas blancas, 20 bolas rojas y 30 bolas verdes. Si tomamos 5 bolas de la caja reemplazándolas (tomando una bola, anotando el color y luego regresandola a la caja). Queremos saber la probabilidad de:

1. Tomar 3 blancas y 2 rojas.
2. Toas todas del mismo color.

Ejecuta el experimento 1000 veces y calcula las probabilidades que mencionadas.

```python
# TODO

import numpy as np

# Definimos la fución para tomar 5 bolas de la caja
def tomar_bolas():
    bolas = ['blanca'] * 10 + ['roja'] * 20 + ['verde'] * 30
    muestra = np.random.choice(bolas, size=5, replace=True)
    return muestra

# Contador para los resultados
contador_3b_2r = 0
contador_mismo_color = 0

num_experimentos = 1000

# Simulación
for _ in range(num_experimentos):
    muestra = tomar_bolas()
    if np.sum(muestra == 'blanca)') == 3 and np.sum(muestra == 'roja') == 2:
        contador_3b_2r += 1
    if np.all(muestra == 'blanca') or np.all(muestra == 'roja') or np.all(muestra == 'verde'):
        contador_mismo_color += 1

# Calculamos la probabilidad
prob_3b_2r = contador_3b_2r / num_experimentos
prob_mismo_color = contador_mismo_color / num_experimentos

#print
print(" La probabilidad de tomar 3 bolas blancas y 2 rojas:", prob_3b_2r)
print("La probabilidad de tomar todas las bolas del mismo color:", prob_mismo_color)
```
